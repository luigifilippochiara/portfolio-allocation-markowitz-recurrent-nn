# Data import

```python
prices = pd.read_csv('dati.csv')
```

```python
# convert to date format
prices['dates'] = pd.to_datetime(prices['dates'])
# set dates as index
prices.set_index('dates', inplace=True)
# filter dates before december, which is incomplete
dataFilter = prices.index < datetime(2018,12,1)
prices = prices[dataFilter]
```

```python
# update column names
prices.columns = [r.split(' ')[0] for r in prices.columns]
prices = prices.dropna()
# remove greece
prices = prices.drop(['GR'], 1)
prices.head()
```

```python
returns = ((prices/prices.shift(1))-1)*100
returns = returns.dropna()
returns.head()
```

# Create training, validation and test sets

```python
# lookback months
windowSize = 60
```

```python
# function to generate input vectors
def generateWindows(data, dates, windowSize = 7, step=1):
    windowSize += 1
    x = []
    y = []
    yDates = []
    denorm = []
    for i in range(0, data.shape[0] - windowSize+1, step):
        dataWin = data[i: i + windowSize,:]
        dataMin = dataWin.min(axis=0)
        dataMax = dataWin.max(axis=0)
        denorm.append([dataMin, dataMax])
        # normalization [0,1]
        dataWin = (dataWin - dataMin) / (dataMax - dataMin)
        x.append(dataWin[:-1])
        y.append(dataWin[-1])
        yDates.append(dates[i: i + windowSize][-1])
    return np.array(x), np.array(y),np.array(denorm),yDates
```

```python
# extract input vectors
x,y,denorm,dates = generateWindows(returns.values, returns.index, windowSize)
```
